```python
from pydub import AudioSegment
import math
import os

def split_audio(file_path, output_folder, clip_length=2100):
    """
    Splits the audio file into clips of specified length.

    :param file_path: Path to the input audio file.
    :param output_folder: Directory to save the output clips.
    :param clip_length: Length of each clip in milliseconds.
    """
    # Load the audio file
    audio = AudioSegment.from_file(file_path)
    total_length = len(audio)
    num_clips = math.ceil(total_length / clip_length)

    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    for i in range(num_clips):
        start_time = i * clip_length
        end_time = min((i + 1) * clip_length, total_length)
        clip = audio[start_time:end_time]
        output_file = os.path.join(output_folder, f"clip_{i+1}.mp3")
        clip.export(output_file, format="mp3")
        print(f"Exported {output_file}")

# Example usage
if __name__ == "__main__":
    input_file = "/content/drive/MyDrive/Speech data/audio_dataset/Elon_Musk/Elon_Musk_raw_2.mp3.mp3"  # Replace with your audio file path
    output_dir = "/content/drive/MyDrive/Speech data/separated_voices/Elon_Musk/chunks"
    split_audio(input_file, output_dir)
```

```python
import os
import glob
import whisper
from pydub import AudioSegment
from phonemizer.phonemize import phonemize

# Paths
input_folder = '/content/drive/MyDrive/Speech data/Elon_musk_seperated'
output_folder = '/content/drive/MyDrive/Speech data/Elon chunked'
os.makedirs(output_folder, exist_ok=True)

# Initialize Whisper model
model = whisper.load_model('base')  # You can use 'tiny', 'small', 'medium', 'large' models

# Function to split audio into chunks of 20-30 words
def split_audio(audio_file, transcription, words_per_chunk=(20, 30)):
    chunks = []
    current_chunk = []
    current_duration = 0
    current_text = ''
    current_start_time = None
    current_end_time = None

    words = transcription['segments']

    for word_info in words:
        word = word_info['text'].strip()
        if not word:
            continue
        if current_start_time is None:
            current_start_time = word_info['start']

        current_chunk.append(word)
        current_end_time = word_info['end']

        if len(current_chunk) >= words_per_chunk[1]:
            chunks.append({
                'start': current_start_time,
                'end': current_end_time,
                'text': ' '.join(current_chunk)
            })
            current_chunk = []
            current_start_time = None
            current_end_time = None

    # Add the last chunk if it's within the specified range
    if words_per_chunk[0] <= len(current_chunk) <= (words_per_chunk[1]+1):
        chunks.append({
            'start': current_start_time,
            'end': current_end_time,
            'text': ' '.join(current_chunk)
        })
    elif len(current_chunk) <= words_per_chunk[0]:
        chunks.append({
            'start': current_start_time,
            'end': current_end_time,
            'text': ' '.join(current_chunk)
        })
    return chunks

# Prepare the final transcription file
transcription_file = os.path.join(output_folder, 'transcriptions.txt')
english_transcription_file = os.path.join(output_folder, 'english_transcriptions.txt')
english_list=[]
with open(transcription_file, 'w', encoding='utf-8') as f_transcript:

    # Process all MP3 files in the input folder
    for mp3_file in glob.glob(os.path.join(input_folder, '*.wav')):
        base_filename = os.path.splitext(os.path.basename(mp3_file))[0]
        print(f'Processing file: {mp3_file}')

        # Load the MP3 file
        audio = AudioSegment.from_file(mp3_file)

        # Save as WAV for Whisper processing
        wav_filename = os.path.join(output_folder, f'{base_filename}.wav')
        audio.export(wav_filename, format='wav')

        # Transcribe using Whisper with word-level timestamps
        result = model.transcribe(wav_filename, word_timestamps=True)
        words = []
        for segment in result['segments']:
            for word_info in segment['words']:
                words.append({
                    'start': word_info['start'],
                    'end': word_info['end'],
                    'text': word_info['word']
                })

        # Re-structure transcription with word-level data
        transcription = {'segments': words}

        # Split the audio into chunks of 20-30 words
        chunks = split_audio(wav_filename, transcription)

        # Load the audio file again for chunking
        audio = AudioSegment.from_wav(wav_filename)

        # Process each chunk
        for idx, chunk in enumerate(chunks):
            chunk_filename = f'{base_filename}_{idx:04d}.wav'
            chunk_path = os.path.join(output_folder, chunk_filename)

            if chunk is None or chunk['start'] is None:
                print(f'Skipping empty chunk: {chunk_filename}')
                continue

            # Extract the audio segment
            start_ms = chunk['start'] * 1000  # Convert to milliseconds
            end_ms = chunk['end'] * 1000
            audio_chunk = audio[start_ms:end_ms]
            audio_chunk.export(chunk_path, format='wav')

            # Convert text to phonemes
            text = chunk['text']
            phonemes = phonemize(
                text,
                language='en-us',
                backend='espeak',
                strip=True,
                preserve_punctuation=False,
                with_stress=False,
                njobs=4
            )

            # Write to the transcription file
            transcription_line = f'{chunk_filename}|{phonemes}|0\n'
            english_transcription_line = f'{chunk_filename}|{text}\n'
            english_list.append(english_transcription_line)
            f_transcript.write(transcription_line)

            print(f'Saved chunk: {chunk_filename}')

with open(english_transcription_file, 'w', encoding='utf-8') as f2_transcript:
    for line in english_list:
        f2_transcript.write(line)

print('Processing complete.')
```

```python
import numpy as np

def increase_volume(audio_data, factor=1.5):
  """Increases the volume of audio data by a given factor.

  Args:
    audio_data: A NumPy array representing the audio data.
    factor: The factor by which to increase the volume (e.g., 1.5 for 50% increase).

  Returns:
    A NumPy array representing the audio data with increased volume.
  """
  return np.clip(audio_data * factor, -1.0, 1.0)
```

```python
def cut_audio(audio_data, start_cut_seconds=0.5, end_cut_seconds=0.5, sample_rate=24000):
  """Cuts the start and end of an audio array.

  Args:
    audio_data: A NumPy array representing the audio data.
    start_cut_seconds: The number of seconds to cut from the start.
    end_cut_seconds: The number of seconds to cut from the end.
    sample_rate: The sample rate of the audio data.

  Returns:
    A NumPy array representing the audio data with the start and end cut.
  """
  start_cut_samples = int(start_cut_seconds * sample_rate)
  end_cut_samples = int(end_cut_seconds * sample_rate)
  return audio_data[start_cut_samples:-end_cut_samples]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import random

def sample_text_based_on_cosine_similarity(input_text, text_list):
  """Samples a text from a list based on cosine similarity with an input text.

  Args:
    input_text: The input text to compare against.
    text_list: A list of strings representing the texts to sample from.

  Returns:
    A randomly sampled text from the list, weighted by cosine similarity.
  """

  # Create a TF-IDF vectorizer
  vectorizer = TfidfVectorizer()

  # Fit and transform the text list and input text
  tfidf_matrix = vectorizer.fit_transform(text_list + [input_text])

  # Calculate cosine similarity between input text and all texts in the list
  cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

  # Replace infinity and NaN values with 0
  cosine_similarities = np.nan_to_num(cosine_similarities, nan=0.0, posinf=0.0, neginf=0.0)

  # Handle the case where all similarities are zero
  if np.sum(cosine_similarities) == 0:
      print("Completely Outside the Distribution, choosing random reference.")
      # If all similarities are zero, return a random text
      return random.choice(text_list)

  # Create a probability distribution based on cosine similarity
  probabilities = cosine_similarities / cosine_similarities.sum()

  # Randomly sample a text from the list using the probability distribution
  sampled_index = random.choices(range(len(text_list)), weights=probabilities)[0]
  return text_list[sampled_index]
```

```python

def process_text_file(file_path):
  """Reads a text file, splits each line by '|', and returns a list of the last elements."""
  with open(file_path, 'r') as f:
    lines = f.readlines()
  file_list = []
  processed_list = []
  for line in lines:
    parts = line.strip().split('|')
    if parts:
      file_list.append(parts[0])
      processed_list.append(parts[-1])
  return file_list,processed_list


#file_path = '/content/drive/MyDrive/Independent Study/StyleTTS2_setup/StyleTTS2/Data/train_test_val_list.txt'
file_path = '/content/drive/MyDrive/Speech data/Elon chunked/english_transcriptions.txt'
file_list,text_list = process_text_file(file_path)
```

```python
import whisperx

def get_word_timestamps_whisperx(audio_path, device):
    # Load the audio
    audio = whisperx.load_audio(audio_path)

    # Load the WhisperX model
    model = whisperx.load_model("small", device=device)

    # Perform noise reduction
    reduced_noise = nr.reduce_noise(y=audio, sr=16000)

    # Transcribe with word-level timestamps
    result = model.transcribe(reduced_noise, batch_size=16, language='en')

    # Load alignment model and metadata
    model_a, metadata = whisperx.load_align_model(language_code='en', device=device)

    # Align Whisper output
    result_aligned = whisperx.align(result["segments"], model_a, metadata, audio, device)
    # result_aligned["word_segments"] contains the word-level timestamps

    words = []
    for word_info in result_aligned["word_segments"]:
        #print(word_info)
        word = word_info['word']
        start_time = word_info.get('start','')
        end_time = word_info.get('end','')
        words.append({
            'word': word,
            'start_time': start_time,
            'end_time': end_time
        })
    return words
```

```python
# Read all file names in a folder and get the numbers in their names into a list

import os
import re

def get_numbers_from_filenames(folder_path):
  """Reads all filenames in a folder and extracts the numbers from them into a list.

  Args:
    folder_path: The path to the folder containing the files.

  Returns:
    A list of integers extracted from the filenames.
  """
  numbers = []
  for filename in os.listdir(folder_path):
    # Extract numbers using a regular expression
    match = re.findall(r'\d+', filename)
    if match:
      numbers.append(int(match[0]))
  return numbers

# Example usage:
folder_path = '/content/drive/MyDrive/Independent Study/Perf_metrics/resampled_ground_truth/resampled_ground_truth'  # Replace with your folder path
numbers_list = get_numbers_from_filenames(folder_path)
#print(numbers_list)
```

```python
# Initialize lists to store the metrics
pesq_scores = []
stoi_scores = []
llr_scores = []
wss_scores = []
snrseg_scores = []

# Initialize lists to store the composite scores
csig_scores = []
cbak_scores = []
covl_scores = []

resampled_ground_truth_dir = '/content/drive/MyDrive/Independent Study/Perf_metrics/resampled_ground_truth/resampled_ground_truth'
resampled_synthesized_dir = '/content/drive/MyDrive/Independent Study/Perf_metrics/resampled_synthesized'

num_samples = len(numbers_list)

for i in tqdm(numbers_list):
    # Paths to resampled audio
    gt_path = os.path.join(resampled_ground_truth_dir, f'gt_{i}.wav')
    synth_path = os.path.join(resampled_synthesized_dir, f'synth_{i}.wav')

    # Load audio
    gt_audio, _ = librosa.load(gt_path, sr=16000)
    synth_audio, _ = librosa.load(synth_path, sr=16000)

    # Get word-level timestamps
    # alignment_gt = get_word_timestamps_whisper(gt_path, model, processor)
    # alignment_synth = get_word_timestamps_whisper(synth_path, model, processor)
    alignment_gt = get_word_timestamps_whisperx(gt_path, device)
    alignment_synth = get_word_timestamps_whisperx(synth_path, device)
    gt_words = [word["word"] for word in alignment_gt]
    synth_words = [word["word"] for word in alignment_synth]
    # Check if the number of words is the same
    if len(alignment_gt) != len(alignment_synth):
        print(f"Mismatch in number of words for sample {i}")
        #continue

    # Initialize per-word metric lists
    per_word_pesq_scores = []
    per_word_stoi_scores = []
    per_word_llr_scores = []
    per_word_csig_score = []
    per_word_wss_scores = []
    per_word_cbak_score = []
    per_word_snrseg_scores = []
    per_word_covl_score = []

    for word_idx in range(len(alignment_gt)):
        word_info_gt = alignment_gt[word_idx]
        synth_index = word_idx
        if gt_words[word_idx] not in synth_words:
            continue
        else:
            if word_idx >= len(alignment_synth):
                continue
            if gt_words[word_idx] != synth_words[word_idx]:
                synth_index = synth_words.index(gt_words[word_idx])
        word_info_synth = alignment_synth[synth_index]

        # Get start and end times
        start_time_gt = word_info_gt['start_time']
        end_time_gt = word_info_gt['end_time']

        start_time_synth = word_info_synth['start_time']
        end_time_synth = word_info_synth['end_time']

        if (start_time_gt == '') or (end_time_gt == '') or (start_time_synth == '') or (end_time_synth == ''):
            continue
        # Convert times to sample indices
        start_sample_gt = int(start_time_gt * 16000)
        end_sample_gt = int(end_time_gt * 16000)

        start_sample_synth = int(start_time_synth * 16000)
        end_sample_synth = int(end_time_synth * 16000)

        # Extract audio segments
        gt_audio_segment = gt_audio[start_sample_gt:end_sample_gt]
        synth_audio_segment = synth_audio[start_sample_synth:end_sample_synth]

        # Ensure same length
        min_length = min(len(gt_audio_segment), len(synth_audio_segment))
        gt_audio_segment = gt_audio_segment[:min_length]
        synth_audio_segment = synth_audio_segment[:min_length]

        # Check if the segment length is sufficient
        if len(gt_audio_segment) < 1600:  # Less than 0.1 seconds
            continue
        # Calculate scores
        pesq_score = pysepm.pesq(gt_audio_segment, synth_audio_segment)[-1]
        stoi_score = pysepm.stoi(gt_audio_segment, synth_audio_segment)
        [csig_score,cbak_score,covl_score] = pysepm.composite(gt_audio_segment, synth_audio_segment)

        per_word_pesq_scores.append(pesq_score)
        per_word_stoi_scores.append(stoi_score)
        per_word_csig_score.append(stoi_score)
        per_word_cbak_score.append(stoi_score)
        per_word_covl_score.append(stoi_score)

        # Compute LLR
        llr_score = pysepm.llr(gt_audio_segment, synth_audio_segment, fs=16000)
        per_word_llr_scores.append(llr_score)

        # Compute WSS
        wss_score = pysepm.wss(gt_audio_segment, synth_audio_segment, fs=16000)
        per_word_wss_scores.append(wss_score)

        # Compute Segmental SNR
        snrseg_score = pysepm.SNRseg(gt_audio_segment, synth_audio_segment, fs=16000)
        per_word_snrseg_scores.append(snrseg_score)

    # Compute mean metrics for the sample
    if per_word_pesq_scores:
        pesq_scores.append(np.mean(per_word_pesq_scores))
    if per_word_stoi_scores:
        stoi_scores.append(np.mean(per_word_stoi_scores))
    if per_word_csig_score:
        csig_scores.append(np.mean(per_word_csig_score))
    if per_word_cbak_score:
        cbak_scores.append(np.mean(per_word_cbak_score))
    if per_word_covl_score:
        covl_scores.append(np.mean(per_word_covl_score))
    if per_word_llr_scores:
        llr_scores.append(np.mean(per_word_llr_scores))
    if per_word_wss_scores:
        wss_scores.append(np.mean(per_word_wss_scores))
    if per_word_snrseg_scores:
        snrseg_scores.append(np.mean(per_word_snrseg_scores))

# Compile results
results = {
    'PESQ': pesq_scores,
    'STOI': stoi_scores,
    'LLR': llr_scores,
    'WSS': wss_scores,
    'SNRseg': snrseg_scores,
    'CSIG': csig_scores,
    'CBAK': cbak_scores,
    'COVL': covl_scores
}

df = pd.DataFrame(results)
df.to_csv('/content/evaluation_results.csv', index=False)

# Compute and print mean and standard deviation
metrics = ['PESQ', 'STOI', 'CSIG', 'CBAK', 'COVL']
for metric in metrics:
    if results[metric]:
        mean_score = np.mean(results[metric])
        std_score = np.std(results[metric])
        print(f'{metric}: Mean = {mean_score:.4f}, Std = {std_score:.4f}')
    else:
        print(f'{metric}: No valid scores computed.')
```
